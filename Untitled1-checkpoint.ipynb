```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
Africa = (world[world['continent'] ==  'Africa'])
gdf1 = gpd.GeoDataFrame(pd.read_csv('GPS_1.csv', encoding='latin-1'), geometry=gpd.points_from_xy(pd.read_csv('GPS_1.csv', encoding='latin-1').Long, pd.read_csv('GPS_1.csv', encoding='latin-1').Lat))
ax = gplt.webmap(Africa, projection=gcrs.WebMercator())
gplt.pointplot(
    gdf1, projection=gcrs.AlbersEqualArea(),
    scale='Number', limits=(4, 50),legend=True,legend_kwargs={'bbox_to_anchor': (1, -0.05), 'frameon': False},
    ax=ax
)
fig = plt.gcf()
fig.set_size_inches(10, 8)
plt.savefig('PPNumb1_.png', bbox_inches = 'tight', dpi=300)
```
